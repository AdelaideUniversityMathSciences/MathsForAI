```python
import torch
import torch.fft as fft
import matplotlib.pyplot as plt
import math
```

```python
fs = 1000
f0 = 100
x0 = 1.0
x1 = 10.0

x = torch.arange(x0, x1, 1/fs, dtype=torch.float32)
N = x.size(dim=0)
y = torch.sin(2*math.pi*f0*x)
z = abs(fft.fftshift(fft.fft(y))) ** 2

f = torch.arange(-fs/2, fs/2, fs/N)
plt.semilogy(f, z)
plt.xlabel('freq (Hz)')
plt.ylim(1.0e-4, 1.0e9)
```

```python
import numpy as np
import IPython
T = 2.0    # seconds
sr = 22050 # sample rate
t = np.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.5*np.sin(2*np.pi*440*t)                # pure sine wave at 440 Hz
IPython.display.Audio(x, rate=sr)
```
