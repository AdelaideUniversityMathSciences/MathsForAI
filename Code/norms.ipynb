# Calculate the level sets of norms

Calcualte the level sets (contours, or level curves) of different vector norms in 2D where we can visualise them.

```python
import torch
import math
```

```python
n = 1000
theta = torch.arange(0.0, 2*math.pi, step=math.pi/n)
print(theta)
x = torch.cos(theta)
y = torch.sin(theta)
```

```python
import matplotlib.pyplot as plt
torch.Tensor.ndim = property(lambda self: len(self.shape)) 
plt.plot(x, y, label='L_2')
plt.axes().set_aspect('equal')
plt.show()
```

```python
ps = [0.25, 0.5, 1.001, 2.0, 3.0, 10.0, 100.0]
for p in ps:
  r = ( 1 / (torch.abs(x) ** p + torch.abs(y) ** p) )**(1/p)
  plt.plot( r*x, r*y, label=f'p = {p}')
plt.axes().set_aspect('equal')
plt.legend(bbox_to_anchor=(1.1, 1.00))
plt.show()
```
