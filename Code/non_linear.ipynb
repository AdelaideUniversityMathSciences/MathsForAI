```python
import torch
import math
import numpy as np
```

```python
n = 10000
Z = torch.zeros(2,n)
A = torch.normal( Z, 1.0 )
theta = 2*np.pi*torch.rand(n)
r = (8 + torch.normal( Z[1,:], 1.0 )).repeat(2,1)
x1 = np.cos(theta)
x2 = np.sin(theta)
B = r * torch.stack( (x1,x2), 0)
```

```python
import matplotlib.pyplot as plt

plt.axes().set_aspect('equal')
plt.scatter(A[0,:], A[1,:], color='red', s=1)
plt.scatter(B[0,:], B[1,:], color='blue', s=2)
plt.show()
```

```python
rA = torch.linalg.vector_norm( A, dim=0, ord=2)
# print(rA)

rB = torch.linalg.vector_norm( B, dim=0, ord=2)
# print(rB)

plt.hist(rA, color='red')
plt.hist(rB, color='blue')
```
